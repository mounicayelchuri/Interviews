#### Task
Provide a script to collect Google Trends weekly, daily and hourly data with the keyword ‘bitcoin’ since 2015-01-01 and save results in csv or json file

##### Approach :
1. Use pytrends api to get the data.
2. Extract the data from January,2015 to a dataframe.
3. Create new attributes like year, month, day,hour, week from date.
4. Group data based on year, month, day and get the sum of counts to get daily data.
5. Group data based on year, month, day, week and get the sum of counts to get weekly data.

```python
# importing libraries
from pytrends.request import TrendReq
import pandas as pd
```

```python
pytrends = TrendReq(hl='en-US', tz=360) 
kw_list = ["bitcoin"]
```

```python
#making the api call per each year and append each year data to the dataframe
df = pd.DataFrame(columns=['bitcoin','isPartial'])
for year in range(2015,2023):
    temp=pytrends.get_historical_interest(kw_list, year_start=year, month_start=1, day_start=1, hour_start=0, year_end=year, month_end=12, day_end=31, hour_end=0, cat=0, sleep=0)
    temp.reset_index();
    df=pd.concat([df,temp])
```

##### few requests are failed with response code 429. response code 429 denotes too many requests in a given time.

```python
df.shape
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
#converting index to datetime object and storing it as date column
df['date'] = pd.to_datetime(df['index'])
```

```python
#dropping unneccessary fields
df.drop(columns=["index","isPartial"], inplace=True)
```

##### Extracting the year, month, day, hour, week columns

```python
df['year']=pd.DatetimeIndex(df['date']).year
```

```python
df['month']=pd.DatetimeIndex(df['date']).month
```

```python
df['day']=pd.DatetimeIndex(df['date']).day
```

```python
df['hour']=pd.DatetimeIndex(df['date']).hour
```

```python
df['week']= df['date'].dt.strftime('%U')
```

```python
df.head(25)
```

```python
df[['year','month','day','hour','bitcoin']].to_csv("hour_data.csv")
```

```python
#grouping the data to get daily counts
daily_counts=df.groupby(['year','month','day']).bitcoin.agg([sum])
```

```python
daily_counts
```

```python
daily_counts.to_csv('daily_data.csv')
```

```python
#grouping the data to get weekly counts
weekly_counts=df.groupby(['year','month','week']).bitcoin.agg([sum])
weekly_counts.to_csv('weekly_data.csv')
```
